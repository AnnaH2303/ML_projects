1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
# Визначаємо символічну змінну
x = sp.symbols('x')

# Визначаємо функцію і її похідну
f = 10 * x - 5
df = sp.diff(f, x)
print(f"Похідна функції 10x-5 дорівнює", df)
```

```python
x = np.linspace(0,10,20)
f = 10 * x - 5
df = 0 * x + 10 # похідна функції df = 10

plt.figure(figsize = (10, 5))
plt.plot(x, f, label = 'f(x) = 10 * x - 5')
plt.plot(x, df, label = 'f(x) = 10')
plt.title('Графік функції f(x) = 10 * x - 5 та її похідної df(x) = 10')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

**При збільшенні кількості годин збільшується оцінка.**

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
# Визначаємо символічну змінну
x = sp.symbols('x')

# Визначаємо функції і їх похідні
f1 = sp.sin(x)
df1 = sp.diff(f1, x)

f2 = sp.ln(x)
df2 = sp.diff(f2, x)

print(f"Похідна функції sin(x) дорівнює {df1}.")
print(f"Похідна функції ln(x) дорівнює {df2}.")
```

```python
x = np.linspace(0,10,20)
f1 = np.sin(x)
df1 = np.cos(x)

plt.figure(figsize = (10, 2))
plt.plot(x, f1, label = 'f(x) = sin(x)')
plt.plot(x, df1, label = 'f(x) = cos(x)')
plt.title('Графік функції f(x) = sin(x) та її похідної df(x) = cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(1,10,20)
f2 = np.log(x)
df2 = 1/x

plt.figure(figsize = (10, 5))
plt.plot(x, f2, label = 'f(x) = ln(x)')
plt.plot(x, df2, label = 'f(x) = 1/x')
plt.title('Графік функції f(x) = ln(x) та її похідної df(x) = 1/x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Визначаємо символічну змінну
x = sp.symbols('x')

# Визначаємо функції і їх похідні
f3 = x**2 - 5 * x + 1
df3 = sp.diff(f3, x)

print(f"Похідна функції f(x) = x**2 - 5 * x + 1 дорівнює {df3}.")
```

```python
x = np.linspace(0,100,100)
f3 = x**2 - 5 * x + 1
df3 = 2 * x - 5

plt.figure(figsize = (10, 5))
plt.plot(x, f3, label = 'f(x) = x**2 - 5 * x + 1')
plt.plot(x, df3, label = 'f(x) = 2 * x - 5')
plt.title('Графік функції f(x) = x**2 - 5 * x + 1 та її похідної df(x) = 2 * x - 5')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
